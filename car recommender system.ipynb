```python
df=df.drop(columns=['weight','model_year','origin'])
```

```python
df['avg']=(df['mpg']+df['cylinders']+df['displacement']+df['horsepower']+df['acceleration'])/5
df
```

```python
df1=df[['name','mpg','cylinders','displacement','horsepower','acceleration']]
```

```python
def recommend(mpg,cylinders,displacement,horsepower,acceleration):
    a=[]
    val=(mpg+cylinders+displacement+horsepower+acceleration)/5
    for i in range(len(df)):
        a.append(abs(val-df['avg'][i]))
    cars_list=sorted(list(enumerate(a)),key=lambda x:x[1])[:5]
    for i in cars_list:
        print(df1.iloc[i[0]])
        print("\n")
```

```python
recommend(18,10,256,100,16)
```
